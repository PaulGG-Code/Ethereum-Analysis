```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('AverageGas.txt' ,header = None)
```

```python
df1.columns = ['date', 'gas']
```

```python
df1.head()
```

```python
df1['date'] = df1['date'].str.strip('()')
```

```python
df1['date'] = df1['date'].str.replace('"','')
```

```python
df1['gas'] = df1['gas'].str.strip('()')
```

```python
df1['gas'] = df1['gas'].astype(float)
```

```python
df1.head()
```

```python
plt.figure(figsize=(20,10))
plt.title('Transaction Gas Price vs Time')
plt.xlabel('Date')
plt.ylabel('Average Gas')
plt.yticks(rotation=90)
plt.xticks(rotation=90)
plt.plot(df1['date'], df1['gas'])
plt.show()
```
